```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("gagnet_losses_3.txt", 'rb') as handle:
    loader = pickle.load(handle)
```

```python
loader
```

```python
train_losses = loader['train_losses']
vall_losses = loader['vall_losses']
```

```python

plt.plot(range(1,len(train_losses)+1),train_losses)
plt.plot(range(1,len(train_losses)+1),vall_losses)
plt.legend(['Train','Validation'])
plt.title('DB-AIAT Loss')
plt.xticks(range(1,len(train_losses)+1, 1))
plt.xlabel('#Epoch')
plt.show()
```
